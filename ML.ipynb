```python
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans

# 1. Generate the same data
X, y = make_blobs(n_samples=500, centers=5, random_state=42)

# 2. Calculate inertia for different cluster numbers (1 to 10)
inertia_values = []
cluster_range = range(1, 11)

for k in cluster_range:
    km = KMeans(n_clusters=k, random_state=2)
    km.fit(X)
    inertia_values.append(km.inertia_)  # km.inertia_ gives the score

# 3. Plot the Elbow curve
plt.plot(cluster_range, inertia_values, marker='o', color='blue')
plt.title("The Elbow Method")
plt.xlabel("Number of Clusters (K)")
plt.ylabel("Inertia")
plt.xticks(cluster_range)
plt.grid(True)
plt.show()
```

```python
.
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
X, y = make_blobs(n_samples=500, centers=5, random_state=42)
km = KMeans(n_clusters=4, random_state=2)
km.fit(X)
labels = km.labels_
centers = km.cluster_centers_
plt.scatter(X[:, 0], X[:, 1], c=labels, s=50, cmap='viridis')
plt.scatter(centers[:, 0], centers[:, 1], c='black', s=200, alpha=0.5)
plt.title("K-Means Clustering")
plt.xlabel("Feature 1")
plt.ylabel("Feature 2")
plt.show()
```

```python
data = china / 255.0 # use 0 ... 1 scale
data = data.reshape(-1, 3)
data.shape
```

```python
def plot_pixels(data, title, colors=None, N=10000):
    if colors is None:
        colors = data

    # Choose a random subset of pixels so the plot runs fast
    rng = np.random.RandomState(0)
    i = rng.permutation(data.shape[0])[:N]
    colors = colors[i]
    R, G, B = data[i].T

    fig, ax = plt.subplots(1, 2, figsize=(16, 6))

    # Red vs Green Plot
    ax[0].scatter(R, G, color=colors, marker='.', alpha=0.5)
    ax[0].set(xlabel='Red', ylabel='Green', xlim=(0, 1), ylim=(0, 1))

    # Red vs Blue Plot
    ax[1].scatter(R, B, color=colors, marker='.', alpha=0.5)
    ax[1].set(xlabel='Red', ylabel='Blue', xlim=(0, 1), ylim=(0, 1))

    fig.suptitle(title, size=20)
    plt.show()
```

```python
# Pass your K-Means cluster centers mapped onto the labels to color-code the scatter plot
plot_pixels(data, title="Input Pixel Space Mapped to 4 K-Means Clusters", colors=centers[labels])
```

```python
# 1. Define the number of clusters (colors) you want to keep
num_colors = 4  # You can change this to 2, 8, 16, etc.
kmeans = KMeans(n_clusters=num_colors, random_state=42)

# 2. Fit the model on your flattened pixel data
kmeans.fit(data)

# 3. Get the cluster assignments for each pixel and the cluster centers (colors)
labels = kmeans.labels_
centers = kmeans.cluster_centers_

# 4. Reconstruct the image by replacing each pixel with its closest cluster center
# We multiply by 255 to scale it back up to the original pixel value range
compressed_data = centers[labels].reshape(320, 320, 3)

# 5. Plot the original and the compressed image side-by-side
plt.figure(figsize=(10, 5))

plt.subplot(1, 2, 1)
plt.imshow(china)
plt.title("Original Image")
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(compressed_data)
plt.title(f"Quantized Image ({num_colors} Colors)")
plt.axis('off')

plt.show()
```
